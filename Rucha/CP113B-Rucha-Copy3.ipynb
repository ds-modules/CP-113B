```python
def scatter_features(feature1, feature2):
    x = df_2000[[feature1]]
    y = df_2000[[feature2]]
    return plt.scatter(x,y)

# create drop down menus for widget
wid_1 = widgets.Dropdown(
    options=df_2000.columns.tolist(),
    description='Variable:',
    disabled=False,
)

wid_2 = widgets.Dropdown(
    options=df_2000.columns.tolist(),
    description='Variable:',
    disabled=False,
)

interact(scatter_features, feature1 = wid_1, feature2 = wid_2)
```

```python
test_color = gdf['area_descr'][10]
```

```python
town_color = re.findall(r'"(.*?)"', test_color)[1]
town_color
```

```python
re.findall(r'California.*',town_color)[0][11:]
```

```python
re.findall(r'.*California',town_color)[0]
```

```python
colors = ['item_0']
towns = ['item_0']

for area in gdf['area_descr']:
    town_and_color = re.findall(r'"(.*?)"', area)[1]
    if len(re.findall(r'California',area))==0:
        if len(re.findall(r'Berkeley',area))>0:
            colors.append(re.findall(r'Berkeley.[^ACD]*',town_and_color)[0][9:])
            towns.append('Berkeley')
        if len(re.findall(r'Oakland',area))>0:
            colors.append(re.findall(r'Oakland.[^ACD]*',town_and_color)[0][8:])
            towns.append('Oakland')
        if len(re.findall(r'San Leandro',area))>0:
            colors.append(re.findall(r'San Leandro.[^ACD]*',town_and_color)[0][11:])
            towns.append('San Leandro')
            
    else: 
        colors.append(re.findall(r'California.[^ACD]*',town_and_color)[0][11:])
        towns.append(re.findall(r'.*California',town_and_color)[0].replace(', California',""))

    
```

```python
gdf['colors'],gdf['towns'] = colors,towns
```
